Os participantes analisarão os dados que investigam o impacto dos choques nos meios de subsistência agrícolas, utilizando o Sistema de Informação de Dados em Emergências (DIEM) da Organização das Nações Unidas para a Alimentação e a Agricultura (FAO).

O DIEM acumula dados continuamente atualizados sobre os impactos agrícolas das crises em mais de 30 países, atingindo 150.000 famílias anualmente através de 37 inquéritos. Estes dados incluem informações sobre cinco categorias: rendimento, culturas, pecuária, segurança alimentar e necessidades de ajuda.

O questionário DIEM abrange cinco áreas temáticas principais:

Rendimentos e choques: Este módulo explora as alterações nos rendimentos e os choques sofridos pelas famílias durante o período recordatório (normalmente três meses). São obtidas informações sobre o tipo e a magnitude da mudança de rendimento no período de referência para as principais fontes de rendimento (agrícolas e não agrícolas) e os principais choques susceptíveis de afectar a capacidade de uma família produzir alimentos ou gerar rendimento. Uma pergunta específica relacionada às restrições do COVID-19 está incluída.
Produção agrícola: O módulo cultura centra-se nas dificuldades encontradas pelos produtores agrícolas e nos seus impactos na produção e nas vendas. Podem incluir acesso limitado a insumos agrícolas e desafios de comercialização. A produção é avaliada como a mudança relativa na área plantada e colhida (alcançada ou projetada) em comparação com um ano típico recente.

Produção pecuária: O módulo pecuária centra-se nas dificuldades encontradas pelos criadores de gado e nos seus efeitos na produção e nas vendas. Estes podem incluir a falta de acesso a rações e outros factores de produção, ou a serviços veterinários, bem como questões de marketing. A produção é medida como uma mudança relativa no tamanho do rebanho em comparação com o mesmo período do ano anterior.
Resultados de segurança alimentar:Os indicadores padrão incluem a Escala de Experiência de Insegurança Alimentar (FIES), a pontuação da diversidade alimentar dos agregados familiares (HDDS) e o índice de estratégias de sobrevivência dos meios de subsistência (LCSI), que são utilizados como indicadores da insegurança alimentar e da perturbação dos meios de subsistência dos agregados familiares.

Necessidade de assistência: Os entrevistados expressam as suas necessidades de assistência a curto e médio prazo em relação aos meios de subsistência agrícolas. Também fornecem informações sobre segurança alimentar e assistência aos meios de subsistência recebidas durante o período recordatório.

#3ª Processamento dos dados

Criação, limpeza e transformação dos dados

```python
covid_df['total_cases'] = covid_df['total_cases'].fillna(0).astype(int)
covid_df['total_deaths'] = covid_df['total_deaths'].fillna(0).astype(int)
covid_df['population'] = covid_df['population'].fillna(0).astype(int)

covid_df['total_cases'] = covid_df['total_cases'].astype(int)
covid_df['total_deaths'] = covid_df['total_deaths'].astype(int)
covid_df['population'] = covid_df['population'].astype(int)

covid_df['%_mortality'] = (covid_df['total_deaths'] / covid_df['total_cases']) * 100
covid_df['%_cases'] = (covid_df['total_cases'] / covid_df['population']) * 100
covid_df['%_deaths'] = (covid_df['total_deaths'] / covid_df['population']) * 100
```

```python
# Agrupe os dados por país
grouped_df = covid_df.groupby('location')
grouped_df
```

```python
# Calcule as maiores taxas de mortalidade, percentual de casos e percentual de mortes para cada país
max_mortality_rates = grouped_df['%_mortality'].max()
max_percent_cases = grouped_df['%_cases'].max()
max_percent_deaths = grouped_df['%_deaths'].max()
```

```python
# Crie um DataFrame com as maiores taxas para cada país
max_rates_df = pd.DataFrame({'Max_Mortality_Rate': max_mortality_rates,
                             'Max_Percent_Cases': max_percent_cases,
                             'Max_Percent_Deaths': max_percent_deaths}).reset_index()
```
